```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ajustes visuais (opcional)
sns.set(style="whitegrid")
# 1) Carregar (coloque o caminho correto se necessrio)
df = pd.read_csv(r"C:\Users\ENG SOFTWARE\Downloads\wine_quality.csv")

# Criar a pasta 'results' se no existir
os.makedirs("results", exist_ok=True)

# 2) Histograma da varivel 'alcohol'
plt.figure(figsize=(8,5))
sns.histplot(df["alcohol"].dropna(), kde=True)
plt.title("Histograma - Alcohol")
plt.xlabel("Alcohol")
plt.ylabel("Frequncia")
plt.tight_layout()
plt.savefig("results/alcohol_hist.png", dpi=150)
plt.show()
plt.show()

# 3) Boxplot de pH segmentado por 'quality'
plt.figure(figsize=(10,6))
sns.boxplot(x="quality", y="pH", data=df)
plt.title("Boxplot de pH por Quality")
plt.xlabel("Quality")
plt.ylabel("pH")
plt.tight_layout()
plt.savefig("results/ph_by_quality_boxplot.png", dpi=150)
plt.show()
```
